```python
import numpy as np
```

```python
#determina si ninguno de los elementos es cero.
a = np.array([1, 2, 3, 4])
respuesta = np.any(a == 0)
print(respuesta)
```

```python
# ejercicio 2 determina si contiene algún elemento que no sea cero.
a = np.array([0, 0, 2, 0])
respuesta = np.any (a!=0)
print (respuesta)
```

```python
#determina si contiene valores que sean infinitos o valores no numéricos (NaN).
a = np.array([1, 2, np.nan, np.inf])
mascara_inf = np.isinf(a)
print(mascara_inf)
mascara_validos = np.isfinite(a)
print(mascara_validos)
```

```python
#4 determina si contiene valores infinitos.
a = np.array([1, 2, np.inf])
respuesta = np.any(np.isinf(a))
print(respuesta)
```

```python
#5 determina si contiene valores NaN.
a = np.array([1, np.nan, 3])
respuesta = np.any (np.isnan(a))
print (respuesta)
```

```python
# 6 realiza una comparación elemento a elemento 
# contemplando mayor, mayor o igual, menor y menor o igual.
del print
a = np.array([1, 3, 5])
b = np.array([2, 3, 4])
mayores = (a > b)
print (mayores)
mayor_o_igual = (a >= b)
print(mayor_o_igual)
menor = (a <= b)
print (menor)


```

```python
#7 determina la cantidad de memoria que ocupa el arreglo.
a = np.array ([1, 7, 13, 105])
memoria = a.nbytes
print (memoria)
```

```python
#8 Crea un arreglo de 10 ceros; uno de 10 unos; y uno de 10 cincos.
del print
cero = np.zeros(10)
print (cero)
uno = np.ones(10)
print (uno)
cinco = np.full (10, 5)
print (cinco)
```

```python
#9 Crea un arreglo que contenga todos los números enteros entre 30 y 70.
a = np.arange(30, 70)
print (a)
```

```python
#10 Crea un arreglo que contenga los números pares entre 30 y 70.
a = np.arange (30, 70, 2)
print (a)
```

```python
#Crea una matriz identidad de 3x3.
a = np.eye (3)
print (a)
```

```python
#12 Genera un número aleatorio entre 0 y 1.
rng = np.random.default_rng()
a = rng.random()
print (a)
```

```python
#13 Crea un arreglo de 15 números aleatorios tomados de una distribución normal estándar.
rng = np.random.default_rng()
a = rng.normal (size=15)
print (a)
```

```python
#14Crea un arreglo con valores entre 15 y 55, e imprime todos los 
# valores excepto el primero y el último.
a = np.arange (15, 55)
sin = np.delete(a, [0, -1])
print (sin)
```

```python
#15 Crea un vector (arreglo) de longitud 10 con
#  valores equitativamente distribuidos entre 5 y 50.
a = np.linspace (5, 50, 10)
print (a)
```

```python
#16 Crea un vector con valores de 0 a 20, y cambia el signo de 
# los números  de 9 a 15.
a = np.arange (0,21)
a[9:16] = a[9:16]*-1
print (a)
```

```python
#17 Convierte el arreglo a = np.array([1, 2, 3]) 
# a una lista de Python.
a = np.array([1, 2, 3]) 
lista = a.tolist()
print (a)
```

```python
#18 Convierte la lista l = [1, 2, 3] a un arreglo de Numpy.
l = [1, 2, 3] 
a = np.array(l)
print (a)
```

```python
#19 Dada la matriz m = np.array([[1, 2], [3, 4]]), verifica qué 
# números son pares y almacena los resultados en otra matriz.
m = np.array([[1, 2], [3, 4]])
pares = (m % 2 == 0)
print (pares)
```

```python
#20 Dado el arreglo m = np.array([[1, 2, 3], [4, 5, 6]]), 
# encuentra el número de filas y columnas de la matriz.
m = np.array([[1, 2, 3], [4, 5, 6]])
filas_y_columnas = m.shape
print (filas_y_columnas)
```
